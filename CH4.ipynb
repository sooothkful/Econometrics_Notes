```python
#4.3
gpareg = smf.ols(formula='colGPA ~ hsGPA + ACT + skipped', data=gpa1)
gpafit = gpareg.fit()
```

```python
gpafit.summary()
```

```python
#CV for alpha = 5% and 1% using t-distribution with 137 df:
alpha = np.array([0.05, 0.01])
cv_t = stats.t.ppf(1-alpha / 2, 137)
cv_t
```

```python
alpha
```

```python
#CV for alpha = 5% and 1% using normal distribution with 137 df:
```

```python
cv_n = stats.norm.ppf(1-alpha / 2)
cv_n
```

```python
gpafit.bse
```

```python
#reproduce t-statistic
tstat = gpafit.params / gpafit.bse
tstat
```

```python
#reproduct p-values
pval = 2 * stats.t.cdf(-abs(tstat), 137)
pval
```

```python
#95% CI:
CI95 = gpafit.conf_int(0.05)
CI95
```

```python
#99% CI
CI99 = gpafit.conf_int(0.01)
CI99
```

```python
#4.3 LINEAR RESTRICTIONS - F TESTS

mlb1 = pd.read_csv("mlb1.csv")
mlb1
```

```python
mlbreg = smf.ols(formula='np.log(salary) ~ years + gamesyr + bavg + hrunsyr + rbisyr', data=mlb1)
mlbfit = mlbreg.fit()
```

```python
mlbfit.summary()
```

```python
mlbfit.params
```

```python
mlbrreg = smf.ols(formula='np.log(salary) ~ years + gamesyr', data=mlb1)
```

```python
mlbrfit = mlbrreg.fit()
mlb_r2_r = mlbrfit.rsquared
mlb_r2_ur =mlbfit.rsquared
```

```python
mlb_r2_ur
```

```python
mlb_r2_r
```

```python
#F-STATISTIC
n = mlb1.shape[0]
fstat = (mlb_r2_ur - mlb_r2_r) / (1 - mlb_r2_ur) * (n-6)/3
```

```python
fstat
```

```python
#CV 
mlb_cv = stats.f.ppf(1-0.01, 3, 347)
mlb_cv
```

```python
#P-value = 1-cdf of appropriate F distribution
fpval = 1-stats.f.cdf(fstat, 3, 347)
fpval
```
