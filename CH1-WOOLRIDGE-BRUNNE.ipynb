```python
#MAKE OUR OWN CLASS - LOCAL VARIABLES ARE ALWAYS INITIATED BY THE __init__ method in Python
c = np.array([[3,6,1], [2,7,4]])
d = np.array([[1,8,6], [3,5,8], [1,1,2]])
```

```python
c
```

```python
d
```

```python
#NUMPY WAY TO MULTIPLY MATRICES
c.dot(d)
```

```python
class myMatrices:
    def __init__(self, C, D):
        self.C = C
        self.D = D
    
    def mult(self):
        N = self.C.shape[0] #number of rows in C
        K = self.D.shape[1] #number of cols in D
        out = np.empty((N,K)) # initialize output
        for i in range(N):
            for j in range(K):
                out[i,j] = sum(self.C[i, :] * self.D[:,j])
        return out
    
```

```python
# create object    
test = myMatrices(c, d)
```

```python

# access local variables
test.C
```

```python
test.D
```

```python
test.mult()
```

```python
class myMatNew(myMatrices):
    def getTotalElem(self):
        N = self.C.shape[0] # num of rows in C
        K = self.D.shape[1] # num of rows in D
        return N * K
```

```python
testsb = myMatNew(c,d)
```

```python
testsb.getTotalElem()
```

```python
myMatNew(c,d).getTotalElem()
```
