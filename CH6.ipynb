```python
#6.1
hprice2 = pd.read_csv("hprice2.csv")
hprice2
```

```python
hprice2reg = smf.ols(formula = 'price ~ nox + crime + rooms + dist + stratio', data = hprice2)
hprice2fit = hprice2reg.fit()
```

```python
hprice2fit.params
```

```python
hprice2fit.summary()
```

```python
#define a function for the standardization:
def scale(x):
    x_mean = np.mean(x)
    x_var = np.var(x, ddof = 1)
    x_scaled = (x-x_mean) / np.sqrt(x_var)
    return x_scaled


```

```python
# standardize and estimate:
hprice2['price_sc'] = scale(hprice2['price'])
```

```python
hprice2
```

```python
hprice2['nox_sc'] = scale(hprice2['nox'])
hprice2['crime_sc'] = scale(hprice2['crime'])
hprice2['rooms_sc'] = scale(hprice2['rooms'])
hprice2['dist_sc'] = scale(hprice2['dist'])
hprice2['stratio_sc'] = scale(hprice2['stratio'])
```

```python
hprice2
```

```python
hprice2regst = smf.ols(formula = 'price_sc ~ 0 + nox_sc + crime_sc + rooms_sc + dist_sc + stratio_sc', data=hprice2)
hprice2regst_results = hprice2regst.fit()
hprice2regst_results.params
```

```python
hprice2regst_results.summary()
```

```python
# "a 1 standard deviation increase in nox decreases price by 0.34 standard deviations."
```
