```python
#7.3 WOOLRIDGE/BRUNNER
# CAN'T FIND THE RIGHT DATASET FOR THE PROBLEM IN BOOK
cps = pd.read_csv("cps78_85.csv")
cps
```

```python

#IN THE BOOK HE TAKES USES CROSSTAB TO CREATE A TABLE OF CATEGORIES AND FREQUENCIES FOR 2 CATEGORICAL VARIABLES:
freq_gender = pd.crosstab(cps['female'], columns='count')
freq_gender
```

```python
freq_union = pd.crosstab(cps['union'], columns='count')
freq_union
```

```python
cpsregC = smf.ols(formula = 'lwage ~ exper + y85educ + C(female) + C(union)', data = cps)
cpsfitC = cpsregC.fit()
cpsfitC.summary()
```

```python
#CAN USE SECOND ARGUMENT IN C FUNCTION TO SPECIFY TREATMENT GROUPS:

ncpsregC = smf.ols(formula = 'lwage ~ exper + y85educ + C(female, Treatment(0)) + C(union, Treatment(1))', data = cps)
ncpsfitC = ncpsregC.fit()
ncpsfitC.summary()

# IN THE ABOVE EQUATION THE C FUNCTION WITH TREATMENT COMMAND WILL SHOW UP IN REGRESSION TABLE - CAN THEN EASILY SEE THAT 
# THE UNION VARIABLE IS CAPTURED WITH C(union)[T.1] dummy variable and that gender is captured in C(female)[T.0] dummy variable.
```
